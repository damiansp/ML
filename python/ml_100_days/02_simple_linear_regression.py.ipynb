```python
import ssl

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

ssl._create_default_https_context = ssl._create_unverified_context
```

```python
data = pd.read_csv(
    'https://raw.githubusercontent.com/Avik-Jain/100-Days-Of-ML-Code/'
    'master/datasets/studentscores.csv')
print(data.shape)
data.head()
```

```python
X = data.Hours.to_numpy().reshape(-1, 1)
y = data.Scores

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=1)
```

```python
regr = LinearRegression()
regr.fit(X_train, y_train)
```

```python
preds = regr.predict(X_test)
```

```python
plt.scatter(X_train, y_train, color='r', label='data')
plt.plot(X_train, regr.predict(X_train), color='b', label='fitted')
plt.legend();
```

```python
plt.scatter(X_test, y_test, color='r', label='test data')
plt.plot(X_test, preds, color='b', label='predictions')
plt.legend();
```
