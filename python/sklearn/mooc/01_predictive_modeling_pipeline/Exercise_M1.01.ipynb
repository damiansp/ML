```python
import ssl

import pandas as pd
import seaborn as sns

ssl._create_default_https_context = ssl._create_unverified_context
```

```python
DATA = (
    'https://raw.githubusercontent.com/INRIA/scikit-learn-mooc/main/'
    'datasets/')
penguins = pd.read_csv(f'{DATA}/penguins_classification.csv')
print(penguins.shape)
penguins.head()
```

```python
penguins.columns = ['clen', 'cdep', 'species']
```

```python
penguins.dtypes
```

```python
penguins.species.value_counts()
```

```python
penguins.hist();
```

```python
sns.pairplot(
    penguins,
    hue='species',
    vars=['clen', 'cdep'],
    plot_kws={'alpha': 0.4});
```
