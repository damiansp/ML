```python
import ssl

import pandas as pd
from sklearn.compose import make_column_selector as selector
from sklearn.preprocessing import OrdinalEncoder

ssl._create_default_https_context = ssl._create_unverified_context
```

```python
df = pd.read_csv(
    'https://raw.githubusercontent.com/INRIA/scikit-learn-mooc/main/'
    'datasets/adult-census.csv')
print(df.shape)
df.head()
```

```python
X = df.drop(columns=['education-num'])
y = X.pop('class')
```

```python
X['native-country'].value_counts()
```

```python
X.dtypes
```

```python
categorical_selector = selector(dtype_include=object)
categoricals = categorical_selector(X)
categoricals
```

```python
X_cat = X[categoricals]
X_cat.head()
```

```python
encoder = OrdinalEncoder().set_output(transform='pandas')
ed_encoded = encoder.fit_transform(X_cat[['education']])
ed_encoded
```

```python
encoder.categories_
```

```python
X_enc = encoder.fit_transform(X_cat)
X_enc[:5]
```
