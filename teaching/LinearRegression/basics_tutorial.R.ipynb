### R Modeling: Syntactic Candy
The general form for all models in R is
```
f(y ~ x1 + x2 + ...., data=dataset)
```
Where y is the _response variable_ (what you are trying to predict) and the _xs_ are the predictors (explanatory variables)
e.g.: 
```
lm(Ozone ~ Solar.R + Wind + Temp + Month + Date + Weekday,
   data=airquality)```

````r
# Use all predictors (other than your y):
all.single <- lm(Ozone ~ ., airquality) # as above
summary(all.single)
````

````r
# Interactions are made with ':'
some.interactions <- lm(
  Ozone ~ Solar.R + Wind + Temp + Month + Date + Weekday + Temp:Date,
  airquality)
summary(some.interactions)
````

````r
interaction.shorthand <- lm(Ozone ~ Solar.R * Wind + Temp * Date + Month,
                          airquality)
summary(interaction.shorthand)
````

````r
all.2way.interactions <- lm(
    Ozone ~ (Solar.R + Wind + Temp + Month + Date + Weekday)^2, 
    airquality)
summary(all.2way.interactions)
````

````r
model.update <- update(all.2way.interactions, . ~ . -Month:Weekday)
summary(model.update)
````

````r
# Because ^ has a special meaning in the context of model calls, to 
# exponentiate, you must use the identity function I()
exponents <- lm(Ozone ~ Temp + Wind + I(Temp^2) + I(Wind^2), airquality)
summary(exponents)
````

## Comparing Models -- Which model is "best"?
* Traditionally
  * Goodness of Fit Measures (R squared and Adjusted R squared)
  * Information Criteria (most commonly the Akaike Information Criterion: AIC; also common is the Bayesian or BIC)
* The Big Data Era
  * Cross-Validation (CV)

### R Squared and Adjusted R Squared:
Pros:
* Intuitive and simple to interpret
* Can be used to compare arbitrary models

Cons: 
* Empirically, they are more prone to overfitting, esp. when used on smaller data sets

https://en.wikipedia.org/wiki/Coefficient_of_determination

````r
simple <- lm(Ozone ~ Temp, airquality)
summary(simple)
````

````r
var(airquality$Ozone, na.rm=T)
plot(airquality$Ozone)
abline(h=mean(airquality$Ozone, na.rm=T))
````

````r
var(resid(simple))
1 - var(resid(simple)) / var(airquality$Ozone, na.rm=T)
plot(resid(simple))
abline(h=0)
````

````r
# Why is R-squared problematic? Why the need for Adj. Rsq?
temp.v <- seq(min(airquality$Temp), max(airquality$Temp), length=100) 
new <- data.frame(Temp=temp.v)
fitted <- predict(simple, newdata=new)

plot(airquality$Ozone ~ airquality$Temp)
lines(temp.v, fitted)
````

````r
quad <- lm(Ozone ~ Temp + I(Temp^2), airquality)
summary(quad)
````

````r
fitted <- predict(quad, newdata=new)

plot(airquality$Ozone ~ airquality$Temp)
lines(temp.v, fitted)
````

````r
str(summary(quad))
````

````r
summary(quad)$r.squared
summary(quad)$adj.r.squared
````

````r
MAX_EXP <- 8
exps <- 1:MAX_EXP
rs <- numeric(MAX_EXP)
adj.rs <- numeric(MAX_EXP)
mods <- list()

plot(airquality$Ozone ~ airquality$Temp)
formula <- 'Ozone ~ '

for (ex in exps) {
  formula <- paste(formula, '+ I(Temp^', ex, ')', sep='')
  mod <- lm(formula=as.formula(formula), airquality)
  rs[ex] <- summary(mod)$r.squared
  adj.rs[ex] <- summary(mod)$adj.r.squared
  mods[[ex]] <- mod
  fitted <- predict(mod, newdata=new)
  lines(temp.v, fitted, col=ex)
}
````

````r
plot(exps, rs, type='l')
lines(exps, adj.rs, col=2)
legend('bottomright', lty=1, col=1:2, legend=c('Rsq', 'Adjusted'))
````

````r
summary(mods[[4]])
````

````r
par(mfrow=c(2, 2))
plot(mods[[4]])
````

````r
# AIC
big.mod <- mods[[8]]
summary(big.mod)
````

````r
reduced <- step(big.mod)
````

````r
summary(reduced)
````

````r
plot(airquality$Ozone ~ airquality$Temp)
fitted <- predict(reduced, newdata=new)
lines(temp.v, fitted, col=2)
````

````r
par(mfrow=c(2, 2))
plot(reduced)
````
